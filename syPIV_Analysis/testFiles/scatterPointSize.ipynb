```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import EllipseCollection
import dask.dataframe as dd
from scipy.special import erf
from itertools import product

columns = ['x0', 'x1', 'x2', 'v0', 'v1', 'v2', 'u0', 'u1', 'u2',
           'a0', 'a1', 'a2', 'av0', 'av1', 'av2', 'au0', 'au1', 'au2',
           'dt', 'cellNum']

filename = "/home/kalagodk/docStuff/xTrack3/output/umData/281/Z281.*"
df = dd.read_csv(filename, delim_whitespace=True, names=columns)

df = df.compute()

df = df.iloc[::100, :3]

# Transform to experimental plane in mm
df["x"] = 25.4 * (2.75 * df.x0 * 1000 - 38.265)
df["y"] = 25.4 * 2.75 * df.x2 * 1000
df["z"] = 25.4 * (2.75 * df.x1 * 1000 - 1.125)

# Filter the experimental plane
df = df[(df.x.between(17.7, 65.03)) & 
        (df.y.between(0.01, 20))]
#print(df.shape)

x = np.linspace(18.191, 61.851, 1280)
y = np.linspace(0.25681, 17.721, 1024)

# Make a projection onto CCD
mfx = 8.6*25.4/43.66
mfy = 6.8*25.4/17.4619
x = x * mfx
y = y * mfy
x, y = np.meshgrid(x, y)

xp = df.x.to_numpy() * mfx # Magnification Factor
yp = df.y.to_numpy() * mfy # Magnification Factor
#print(xp.shape)

radius = 281e-9 / 2 # in m
sx = 2.0
sy = 1.6
# mfx and mfy to magnify the particles
intensity = lambda x, y, xp, yp: np.sum(np.pi/8 * mfx * mfy * \
                   (2*radius)**2 * sx * sy * \
                   (erf((x-xp + 0.5)/(sx * 2**0.5)) - erf((x-xp - 0.5)/(sx * 2**0.5))) * \
                   (erf((y-yp + 0.5)/(sy * 2**0.5)) - erf((y-yp - 0.5)/(sy * 2**0.5))), axis=0)


xc = np.repeat(x[None, :, :], len(xp), axis=0) # array used for computation
yc = np.repeat(y[None, :, :], len(yp), axis=0)
#print(xc.shape, yc.shape)
xpc = (np.ones(xc.shape).T * xp).T
ypc = (np.ones(yc.shape).T * yp).T
I1 = intensity(xc, yc, xpc, ypc)
#print(xc, yc, xpc, ypc)
#print(I1[np.nonzero(I1)])
######################################################################
my_dpi = 96  # dots per inch (DPI)
xsize = 1280/my_dpi
ysize = 1024/my_dpi
fig = plt.figure(figsize=[xsize, ysize], dpi=my_dpi)
#fig.patch.set_facecolor('black')
#fig.patch.set_alpha(1.0)
#plt.figure()
ax1 = plt.axes([0.0, 0.0, 1.0, 1.0], xlim=(x.min(), x.max()),
              ylim=(y.min(), y.max()))
ax1.contourf(x, y, I1, cmap='jet')
#ax1.imshow(I1, cmap='gray', origin='lower')
#ax1.grid()
plt.show()

my_dpi = 96  # dots per inch (DPI)
xsize = 1280/my_dpi
ysize = 1024/my_dpi
fig = plt.figure(figsize=[xsize, ysize], dpi=my_dpi)
ax = plt.axes([0.0, 0.0, 1.0, 1.0])
ax.imshow(I1, cmap='gray', origin='lower')
plt.show()
fig.savefig("test.tif", dpi=my_dpi)

my_dpi = 96  # dots per inch (DPI)
xsize = 1280/my_dpi
ysize = 1024/my_dpi
fig = plt.figure(figsize=[xsize, ysize], dpi=my_dpi)
#fig.patch.set_facecolor('black')
#fig.patch.set_alpha(1.0)
ax = plt.axes([0.0, 0.0, 1.0, 1.0], xlim=(x.min(), x.max()),
              ylim=(y.min(), y.max()))
ax.scatter(xp, yp, color='r')
ax.grid()
```
