```python
import matplotlib.pyplot as plt
import numpy as np
my_dpi = 100
plt.figure(figsize=[500/my_dpi, 400/my_dpi], dpi=my_dpi)
ax = plt.axes([0.1, 0.1, 0.8, 0.8], xlim=(0, 1), ylim=(0, 1))
points_whole_ax = 500/my_dpi * 0.8 * 72    # 1 point = dpi / 72 pixels
points_whole_ay = 400/my_dpi * 0.8 * 72 
radius = 0.1
points_radius_x = 2 * radius / 1.0 * points_whole_ax
points_radius_y = 2 * radius / 1.0 * points_whole_ay
ax.scatter([0.45, 0.5], [0.45, 0.5], s=points_radius_x*points_radius_y,
           color='r', marker=',')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import EllipseCollection

x, y = 1500 * np.random.rand(2, 100)
size = 50 + 10 * np.random.randn(100)
color = np.random.rand(100)
cmap = plt.cm.hsv

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
ax[0].scatter(x, y, s=size, c=color, cmap=cmap)

offsets = list(zip(x, y))
ax[1].add_collection(EllipseCollection(widths=size, heights=size, angles=0, units='xy',
                                       facecolors=plt.cm.hsv(color),
                                       offsets=offsets, transOffset=ax[1].transData))
#ax[1].axis('equal') # set aspect ratio to equal
ax[1].axis([-400, 1800, -200, 1600])
```

```python
from scipy.special import erf
x = np.arange(0, 25.1, 0.1)
y = np.arange(0, 25.1, 0.1)
x, y = np.meshgrid(x, y)
sx = 0.5
sy = 0.5
radius = 1000
intensity = lambda x, y, xp, yp: np.pi/8 * (2*radius)**2 * sx * sy *\
                      (erf((x-xp + 0.5)/(sx * 2**0.5))  - erf((x-xp - 0.5)/(sx * 2**0.5))) * \
                      (erf((y-yp + 0.5)/(sy * 2**0.5))  - erf((y-yp - 0.5)/(sy * 2**0.5)))

I1 = intensity(x, y, 12.5, 12.5)
```

```python
my_dpi = 100
plt.figure(figsize=[5, 5], dpi=my_dpi)

ax = plt.axes([0.1, 0.1, 0.8, 0.8], xlim=(0, 25), ylim=(0, 25))

ax.contourf(x, y, I1, cmap='hot')
plt.grid()

my_dpi = 100
plt.figure(figsize=[5, 5], dpi=my_dpi)

ax = plt.axes([0.1, 0.1, 0.8, 0.8])

ax.imshow(I1, cmap='gray', origin='lower')
```

```python
from scipy.special import erf
x = np.linspace(0, 12, 50)
y = np.linspace(0, 12, 50)
x, y = np.meshgrid(x, y)
np.random.seed(7)
xp = np.random.randint(0, 12, 5)
yp = np.random.randint(0, 12, 5)

radius = 10
sx = 0.9
sy = 0.9
intensity = lambda x, y, xp, yp: np.mean(np.pi/8 * (2*radius)**2 * sx * sy *\
                      (erf((x-xp + 0.5)/(sx * 2**0.5))  - erf((x-xp - 0.5)/(sx * 2**0.5))) * \
                      (erf((y-yp + 0.5)/(sy * 2**0.5))  - erf((y-yp - 0.5)/(sy * 2**0.5))), axis=0)

ax = plt.axes([0.1, 0.1, 0.8, 0.8], xlim=(0, 12), ylim=(0, 12))
ax.scatter(xp, yp, color='r', marker='o')
plt.grid()
plt.show()

xc = np.repeat(x[None, :, :], len(xp), axis=0) # array used for computation
yc = np.repeat(y[None, :, :], len(yp), axis=0)
print(xc.T.shape, xp.shape)
xpc = (np.ones(xc.shape).T * xp).T
ypc = (np.ones(yc.shape).T * yp).T
print(xc.shape, yc.shape, xpc.shape, ypc.shape, xp.shape)

I1 = intensity(xc, yc, xpc, ypc)
print(I1.shape)

plt.figure(figsize=[5, 5], dpi=my_dpi)
plt.contourf(x, y, I1, cmap='gray')
plt.grid()
```
